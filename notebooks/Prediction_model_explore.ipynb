```python
def outcome_categorize(outcome):
    if 'Offender' in outcome:
        return 'court_ruled'
    if 'Defendant' in outcome:
        return 'court_ruled'
    if outcome == 'Suspect charged as part of another case':
        return 'court_ruled'
    if 'not in the public interest' in outcome:
        return 'case proceedings not in the public interest'
    if outcome in ['Action to be taken by another organisation','Local resolution']:
        return 'case found non-judicial resolution'
    if outcome in ['Court case unable to proceed','Unable to prosecute suspect','Investigation complete; no suspect identified']:
        return 'investigation completed, no court proceeding'
    if outcome in ['Awaiting court outcome','Court result unavailable']:
        return 'investigation completed, pending court action'
    if outcome in ['Under investigation']:
        return 'Under investigation'
    if outcome in ['no outcome','Status update unavailable']:
        return 'no investigation'
```

```python
def classif_ordering(classif):
    if classif =='no investigation':
        return 0
    if classif =='Under investigation':
        return 1
    if classif =='investigation completed, pending court action':
        return 2
    if classif =='investigation completed, no court proceeding':
        return 3
    else:
        return 4
```

```python
outcomes_df['order'] = outcomes_df['classification'].apply(lambda x: classif_ordering(x))
outcomes_df = outcomes_df.sort_values(by = ['order'])
outcomes_df
```
