```python
import pandas as pd
```

```python
merged_df = pd.read_csv('../raw_data/merged2.csv')
```

```python
merged_df.head()
```

```python
#Reduce to only the columns that interest us:
merged_df = merged_df.drop(columns = ['Unnamed: 0','Month','Falls within','LSOA code','LSOA name','Context'])
merged_df.head()
```

```python
merged_df['Last outcome category'].fillna('no outcome', inplace = True)
merged_df['Crime ID'].fillna('no id', inplace = True)
merged_df.head()
```

```python
merged_df.info()
```

```python
dtype_change = {'Crime ID':'string','Longitude':'float32','Latitude':'float32','Crime type':'category','Last outcome category':'category'}
dtype_change
```

```python
merged_df.astype(dtype_change).dtypes
```

```python
merged_df = merged_df.astype(dtype_change)
```

```python
merged_df.info()
```

```python
merged_df.columns = ['id','lon','lat','crime','outcome']
```
