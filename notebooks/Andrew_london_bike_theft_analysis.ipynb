### London Bike Theft Analysis 

### Imports 

```python
import pandas as pd 
import numpy as np
import os
import glob
import folium
import folium.plugins
import datetime
```

### Complete Dataset

```python
print(os.getcwd())
```

```python
directory = os.path.join(os.getcwd(),'data','london_data')
directory
```

```python
dfs = []

for root,dirs,files in os.walk('./data/london_data'):
    path = root + '/'+files[0]
    if path.endswith('.csv'):
        dfs.append(pd.read_csv(path))
        
```

```python
master_df = pd.DataFrame()
for i in range(len(dfs)): 
    master_df = pd.concat([master_df,dfs[i]])
```

### Creating a simple heatmap

```python
bike_theft_location = master_df[master_df["Crime type"]=='Bicycle theft'][['Longitude','Latitude']]
```

```python
bike_theft_location=bike_theft_location.dropna()
```

```python
lat_list = bike_theft_location['Latitude'].tolist()
long_list = bike_theft_location['Longitude'].tolist()
bike_theft_coordinates = set(zip(lat_list,long_list))
```

```python
london = (51.507514, -0.116018)
base_map = folium.Map(location=london, zoom_start=10,min_opacity=100)
folium.plugins.HeatMap(bike_theft_coordinates,overlay=True,min_opacity=0.1).add_to(base_map)
display(base_map)
```

### Creating a heatmap with time features

```python
master_df["Time"] = master_df["Month"].map(lambda x:(x).replace('-',''))
```

```python
master_df["Time"] = master_df["Time"].map(lambda x:int(x))
```

```python
simplified_df = master_df[["Longitude","Latitude","Time"]]
```

```python
simplified_df = simplified_df.dropna()
```

```python
simplified_df[simplified_df["Time"]==202105]
```

```python
lat_long_list = []
for i in range(201806,202105):
    temp=[]
    for index,instance in simplified_df[simplified_df['Time'] == i].iterrows():
        temp.append([instance['Latitude'],instance['Longitude']])
    lat_long_list.append(temp)
```
